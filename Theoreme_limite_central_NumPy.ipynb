```python
#Theoreme limite central avec NumPy
#une variable aléatoire qui serait la somme d'autres variables aléatoires a une distribution gautienne.
```

```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt 
from math import exp, sqrt, pi
```

```python
domaine =range(-100, 100)
mu = 0
sigma = 20

f = lambda x : 1/(sqrt(2*pi*pow(sigma,2))) * exp(-pow((x-mu),2)/(2*pow(sigma,2)))

y = [f(x) for x in domaine]

plot = plt.plot(domaine, y)
```

```python
#Pour que ce soit davantage pertinent on va augmenter le nombre de données
#chaque ligne représente une variable aleatoire et 
#chaque colonne représente un échantillon pour la variable aléatoire
matrice = np.random.rand(100,200)
```

```python
#On pourrait lancer une visualisation de la matrice 
#mais elle est un peu grande
#On peut exploiter la fonction 'sum"
#pour avoir la somme totale des données de la matrice
np.sum(matrice)
```

```python
#Pour la somme des lignes
somme = np.sum(matrice, 0)
```

```python
somme.shape
```

```python
#Il y a 200 éléments 
#ON va les traiter comme échantillons des 100 variables aléatoires de notre matrice
plt.scatter(range(200), somme)
```

```python
#On voit une tendance autour de 50. 
#Pour confirmer cela on fait l'histogramme de ce vecteur
plt.hist(somme)
```

```python
#Pour voir encore plus clairement on va générer une matrice avec bcp plus d'éléments

matrice = np.random.rand(1000, 10000)
```
